```python
from twython import Twython
from twython import TwythonStreamer

class MyStreamer(TwythonStreamer):
    def on_success(self, data):
        if 'text' in data:
            string = data['text'].encode('utf-8')
            f = open('streaming.txt', 'a')
            f.write(string.decode('utf-8'))
            print(string.decode('utf-8'))

    def on_error(self, status_code, data):
        print(status_code, data)
        

# Requires Authentication as of Twitter API v1.1
stream = MyStreamer(APP_KEY, APP_SECRET,
                    OAUTH_TOKEN, OAUTH_TOKEN_SECRET)

f = open('streaming.txt', 'w')  ## ì‹¤í–‰í•  ë•Œ ë§ˆë‹¤ text file ì´ˆê¸°í™” ë¨.

words = ['BTC', 'ì½”ì¸', 'ë¹„íŠ¸ì½”ì¸', 'ì´ë”ë¦¬ì›€', 'ETH']
stream.statuses.filter(track=words)
# stream.user()
# Read the authenticated users home timeline
# (what they see on Twitter) in real-time
# stream.site(follow='twitter')
```
